```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\HP\OneDrive\Documents\Crime_Data_from_2020_to_Present.csv")
```

```python
df
```

```python
#number of missing rows 
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
print(df.columns.tolist())
```

```python
df.drop(['weapon desc', 'vict descent', 'rpt dist no', 'crm cd', 'weapon used cd', 
         'crm cd 1', 'crm cd 2', 'crm cd 3', 'crm cd 4', 'cross street'], 
        axis=1, inplace=True)
```

```python
print(df.columns.tolist())
```

```python
df['date rptd'] = pd.to_datetime(df['date rptd'], format='%m/%d/%Y %I:%M:%S %p')
df['date rptd'] = df['date rptd'].dt.date  # Optional: keep only date
```

```python
df['date occ'] = pd.to_datetime(df['date occ'], format='%m/%d/%Y %I:%M:%S %p')
df['date occ'] = df['date occ'].dt.date  # Optional: keep only date
```

```python
# Convert TIME OCC to 4-digit string with leading zeros (e.g., 30 → "0030")
df['time occ'] = df['time occ'].astype(str).str.zfill(4)

# Format it as HH:MM
df['time occ'] = df['time occ'].str[:2] + ':' + df['time occ'].str[2:]
```

```python
# Step 2: Get top 10 most common crimes
top_crimes = df['crm cd desc'].value_counts().head(10)

# Step 3: Display the result
print("Top 10 most common crime types in LA since 2020:")
print(top_crimes)

```

```python
df['date occ'] = pd.to_datetime(df['date occ'], errors='coerce')
```

```python
pre_covid = df[df['date occ'] < '2020-03-01']
post_covid = df[df['date occ'] >= '2020-03-01']

print("Total crimes before COVID:", len(pre_covid))
print("Total crimes after COVID:", len(post_covid))
```

```python
print(df.columns)
```

```python
non_violence_count = df[df['part 1-2'] == 1].shape[0]
violence_count = df[df['part 1-2'] == 2].shape[0]

print("Non-violent crimes count:", non_violence_count)
print("Violent crimes count:", violence_count)
```

```python
df['date'] = pd.to_datetime(df['date rptd'])  # or the correct date column
df['year'] = df['date'].dt.year
```

```python
crime_by_year = df[df['part 1-2'].isin([1, 2])].groupby(['year', 'part 1-2']).size().unstack(fill_value=0)
crime_by_year.columns = ['Non-violent', 'Violent']
print(crime_by_year)
```

```python
df.groupby('year')['part 1-2'].value_counts()
```

```python
df['month_year_str'] = df['date'].dt.strftime('%B %Y')

monthly_crime_table = df.groupby(['month_year_str', 'part 1-2']).size().unstack(fill_value=0)
monthly_crime_table.columns = ['Non-violent', 'Violent'] 

monthly_crime_table['Total Crimes'] = monthly_crime_table['Non-violent'] + monthly_crime_table['Violent']

Top5_crimes = monthly_crime_table['Total Crimes'].sort_values(ascending=False).head(5)
least5_crimes = monthly_crime_table['Total Crimes'].sort_values(ascending=True).head(5)

print("Top 5 crime months:\n", Top5_crimes)
print("\nLeast 5 crime months:\n", least5_crimes)
```

```python
df['weekday'] = df['date'].dt.day_name()

df['weekday_type'] = df['weekday'].apply(
    lambda x: 'Weekday' if x in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 'Weekend'
)

weekday_type_table = df.groupby(['weekday_type', 'part 1-2']).size().unstack(fill_value=0)
weekday_type_table.columns = ['Non-violent', 'Violent']
weekday_type_table['Total_crimes'] = weekday_type_table['Non-violent'] + weekday_type_table['Violent']

print(weekday_type_table['Total_crimes'])

```

```python
df.columns
```

```python
time_occ_table = df.groupby(['time occ', 'crm cd desc']).size().unstack(fill_value=0)

# Print the table
print("Crime counts by time of day and crime type:")
print(time_occ_table)
```

```python
area_total_crimes = df.groupby('area name').size()

area_total_crimes.sort_values(ascending=False).head(5)
```

```python
monthly_area_crime = df.groupby(['area name', 'month_year_str']).size().unstack(fill_value=0)

# Print areas with highest average monthly crime counts (possible hotspots)
hotspots = monthly_area_crime.mean(axis=1).sort_values(ascending=False).head(10)

print(hotspots)
```

```python
df['date rptd'] = pd.to_datetime(df['date rptd'])  # Convert to datetime first
df['month'] = df['date rptd'].dt.strftime('%B')             # Extract the month
```

```python
import matplotlib.pyplot as plt

# Group by year and count crimes
yearly_trend = df.groupby('year').size()

# Plotting
plt.figure(figsize=(10, 5))
yearly_trend.plot(kind='line', marker='o')
plt.title('Yearly Trend of Reported Crimes')
plt.xlabel('Year')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

top_crimes = df['crm cd desc'].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_crimes.plot(kind='bar')  

plt.title('Top 10 Most Common Crimes')
plt.xlabel('Crime Type')    
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)     
plt.tight_layout()
plt.show()
```

```python
df['hour'] = pd.to_datetime(df['time occ'], format='%H:%M').dt.hour
```

```python
import seaborn as sns

heatmap_data = df.groupby(['weekday', 'hour']).size().unstack(fill_value=0)
heatmap_data = heatmap_data.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

plt.figure(figsize=(14, 6))
sns.heatmap(heatmap_data, cmap="YlOrRd", linewidths=0.5)
plt.title('Crime Frequency by Day of Week and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.tight_layout()
plt.show()
```

```python
# Assuming 'part 1-2' column uses 1 = Non-violent, 2 = Violent
crime_counts = df['part 1-2'].value_counts()
labels = ['Non-violent', 'Violent']
colors = ['skyblue', 'salmon']

plt.figure(figsize=(6, 6))
plt.pie(crime_counts, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Violent vs Non-Violent Crime Breakdown')
plt.axis('equal')  # Equal aspect ratio makes the pie chart a circle.
plt.show()
```
